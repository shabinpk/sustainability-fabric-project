```python
import pandas as pd
df_raw = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/raw/owid-co2-data.csv")
df=df_raw.toPandas()
df.head()
```

```python
df_clean=df[(df['year']>=2000) & (df['iso_code'].notna())& (df['population'].notna())].copy()
```

```python
df_clean=df_clean[df_clean['iso_code'].str.len()==3]
df_clean.sample(5)
```

```python
df_clean['co2_intensity']=df_clean['co2']/df_clean['gdp']
df_clean['co2_calc_per_capita']=df_clean['co2']/df_clean['population']
```

```python
df_clean.info()
```

```python
fact_df=df_clean[[ "country",
        "iso_code",
        "year",
        "co2",
        "co2_per_capita",
        "gdp",
        "population",
        "primary_energy_consumption",
        "energy_per_capita",
        "co2_intensity",
        "co2_calc_per_capita"]].copy()
fact_df.head(10)
```

```python
spark.createDataFrame(fact_df).write.mode('overwrite').saveAsTable('fact_climate')
```

```python
fact_df.head()
```

```python
dim_country=fact_df[['country','iso_code']].drop_duplicates().copy()
dim_country=dim_country.rename(columns={'iso_code':'country_id'})
dim_country.head()
```

```python
spark.createDataFrame(dim_country).write.mode('overwrite').saveAsTable('dim_country')
```

```python
dim_year=fact_df[['year']].drop_duplicates().copy()
dim_year['decade']=(dim_year['year']//10)*10
dim_year=dim_year.sort_values('year')
dim_year.head()
```

```python
spark.createDataFrame(dim_year).write.mode('overwrite').saveAsTable('dim_year')
```
